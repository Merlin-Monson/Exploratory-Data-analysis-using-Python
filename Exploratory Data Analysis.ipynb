```python
#Library import
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("TWO_CENTURIES_OF_UM_RACES.csv")
```

```python
#Checking whether the data has been imported
df.head(10)
```

```python
df.shape 
```

```python
df.dtypes
```

```python
#Cleaning data
#Only the USA races ,50km or 50 mi in 2020
#50 km
#50 mi
df[df['Event distance/length']=='50mi']
```

```python
#Combining the data of 50 km and 50 mi using isin, 2020 year, USA event
df[(df['Event distance/length'].isin(['50km','50mi']))& (df['Year of event']==2020)]
```

```python
#USA
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)=='USA']
```

```python
#Combining all the filter together
df[(df['Event distance/length'].isin(['50km','50mi']))& 
(df['Year of event']==2020)&
(df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)=='USA')]
```

```python
df2=df[(df['Event distance/length'].isin(['50km','50mi']))& 
(df['Year of event']==2020)&
(df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)=='USA')]
```

```python
df2.head(10)
df2.shape
```

```python
#removing USA from event name:
df2['Event name']=df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

```python
#Cleaning athlete age (Creating the age by subtracting the birthyear from the event date
df2['athlete_age']=2020-df2['Athlete year of birth']
```

```python
df2.head()
```

```python
#remove h from athlete performance
df2['Athlete performance']=df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head()
```

```python
#Dropping columns:Athlete club, Athlete country,Athlete year of birth,Athlete age category
df2=df2.drop(['Athlete club', 'Athlete country','Athlete year of birth','Athlete age category'],axis=1)
```

```python
df2
```

```python
#Cleaning up null values
df2.isna().sum()
```

```python
df2[df2['athlete_age'].isna()==1]
```

```python
df2=df2.dropna()
df2.shape
```

```python
#Checking for duplicates
df2[df2.duplicated()==1]
```

```python
#Resetting index
df2.reset_index(drop=1)
```

```python
#Fix types
df2.dtypes
```

```python
df2['athlete_age']=df2['athlete_age'].astype(int)
```

```python
df2['Athlete average speed']=df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
#Rename columns :Year of event,Event dates,Event name,Event distance/length,Event number of finishers,Athlete performance	
#Athlete gender,Athlete average speed,Athlete ID,athlete_age

df2=df2.rename(columns={'Year of event':'year',
                        'Event dates':'race_date',
                        'Event name':'race_name',
                        'Event distance/length':'race_length',
                        'Event number of finishers':'race_number_of_finishers',
                        'Athlete performance':'athlete_performance',
                        'Athlete gender':'athlete_gender',
                        'Athlete average speed':'athlete_average_speed',
                        'Athlete ID':'athlete_id'})
```

```python
df2.head()
```

```python
#reordering columns
df3=df2[['race_date','race_name','race_length','race_number_of_finishers','athlete_id','athlete_gender','athlete_age','athlete_performance','athlete_average_speed']]
df3.head()
```

Finding:
1) Difference in speed for the 50km ,50mi male to female
2) What age groups are the best in the 50 mi race (20+races min)
3) What age groups are the worst in the 50 mi race (20+races min)
4) Are runners slower in summer than winter:
   spring:3-5
   summer:6-8
   fall:9-11
   winter:12-2



```python
#Difference in speed for the 50km ,50mi male to female
df3.groupby(['race_length','athlete_gender'])['athlete_average_speed'].mean()
```

```python
#What age groups are the best in the 50 m race (20+races min)
df3.query('race_length=="50mi"') \
   .groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']) \
   .sort_values(by='mean', ascending=False) \
   .query('count>10') \
   .head(20)
```

```python
#What age groups are the worst in the 50 m race (20+races min)
df3.query('race_length=="50mi"') \
   .groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']) \
   .sort_values(by='mean', ascending=True) \
   .query('count>10') \
   .head(20)
```

```python
df3['race_month']=df3['race_date'].str.split('.').str.get(1).astype(int)
```

```python
df3.head(100)
```

```python
df3['race_season']=df3['race_month'].apply(lambda x:'Winter' if x>11 else 'Fall' if x>8 else 'Summer' if x>5 else "Spring" if x>2 else 'Winter')
```

```python
df3.head(100)
```

```python
df3.groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=False)
```

```python
#50 miler only
df3.query('race_length=="50mi"').groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=False)
```
